```python
# selecting Iv & Dv
X = df.iloc[:,[3,11,12,15]]
Y= df.iloc[:,1]
```

```python
# Standardizing data
from sklearn import preprocessing
x = X.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_trans = pd.DataFrame(x_scaled)
```

```python
# Train test split
from sklearn.model_selection import train_test_split as tts
X_train, X_test, Y_train, Y_test = tts(X_trans, Y, test_size=0.25, random_state=0)
```

```python
# Rearranging index for Dataframes
import numpy as np
Y_train.index = np.arange(len(Y_train))
X_train.index = np.arange(len(X_train))
```

```python
# Model Fitting
from sklearn import linear_model as lm
reg = lm.LinearRegression()
reg.fit(X_train,Y_train)
```

```python
# Predicting on test data
Y_pred = reg.predict(X_test)
validate = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
```

```python
# Evaluating
from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
```
